```python
"""=============================================================================
Ex3: Pseudo Inverse Matrix
    Câu 3: 
        a) Tạo ma trận A(4 x 6) chứa các giá trị ngẫu nhiên trong khoảng 1 - 10
        b) Tạo ma trận giả nghịch đảo của A
============================================================================="""
print('\n---------------------Cau 3--------------------------------')
import random
import numpy as np
from numpy.linalg import inv

#a) Create A(4x6) matrix whose values range from 1 to 10 randomly
def cre_matrix(m,n,start,end):
    matrix = []
    for i in range(m):
        row = []
        for j in range(n):
            row.append(random.randint(start,end+1))
        matrix.append(row)
    return np.array(matrix)

A = cre_matrix(4,6,1,10)
print('Matrix A'+str(A.shape)+' =\n',A)

#b) Create pseudo matrix of A
print('\n********************************************************************************************')
print('1st method: create pseudo inverse matrix of A by using numpy.linalg.pinv(A)')
Api = np.linalg.pinv(A)
print('\nPseudo inverse matrix mtrApi of A =',Api)
print('\nCheck if Api.A = In =\n',Api@A)
print('\nCheck if A.Api = Im =\n',A@Api)

print('\n********************************************************************************************')
print('2nd method: using SVD, Api = V.inv(S).Ut to create pseudo inverse matrix')
U,s,Vt = np.linalg.svd(A)

#Do SVD analysis of A
print('\nThe U'+str(U.shape)+' orthogonal matrix of A =\n',U)
print('\nThe s sqrt(eigenvalues) of A =\n',s)
print('\nThe Vt'+str(Vt.shape)+' normalized orthogonal matrix of A =\n',Vt)

#Since S(6x4) will not be a square matrix, so we have to create the inverse S matrix manually by following codes
invS = np.zeros(A.T.shape) #create a zero matrix have the same shape of matrix transpose(A) = 6x4
                           #The shape of Vt is 6x6, so the V is also 6x6
                        #Therefore we have to create a 6x4 invS matrix to be able to multiply the matrix V in V.invS.Ut
                        #That's why we use np.zeros(A.T.shape)

d = 1.0 / s #calculate the inverse of s values
invS[:len(s),:len(s)] = np.diag(d) #len(s) = 4, so this command is to add the s values into first 4 diagonal positons
print('\nThe inverse S matrix of A =\n',invS)

#Api = V.inv(S).Ut
Api = (Vt.T).dot(invS).dot(U.T)

print('\nThe pseudo inverse matrix Api = (Vt.T).dot(inv(S)).dot(U.T) =\n',Api)
print('\nCheck if Api.A = In =\n',Api@A)
print('\nCheck if A.Api = Im =\n',A@Api)
```
