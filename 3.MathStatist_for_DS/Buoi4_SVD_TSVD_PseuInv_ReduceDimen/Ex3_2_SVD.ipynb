```python
"""=============================================================================
Ex3: SVD
    Câu 2: 
        a) Cho tập tin iris.csv, đọc dữ liệu ra dataframe và chỉ lấy 4 cột đầu
        b) Phân tích SVD: U, s, VT từ dataframe
        c) Tạo dataframe mới từ U, s, VT, nhận xét và rút gọn thành phần, nếu có thể
        d) Tìm error nếu có rút gọn thành phần 
        e) Tái cấu trúc dataframe theo U, s, VT (giữ nguyên tất cả các thành phần)
============================================================================="""
print('-----------------------Cau 2-------------------------')
import numpy as np
import pandas as pd

#a) Read the iris.csv file into a dataframe and print out the first 4 rows
df_iris = pd.read_csv('data_Buoi4/iris.csv')
print(df_iris.head(4))

#b) Do SVD analysis U, S, Vt of df_iris
U,S,Vt = np.linalg.svd(df_iris.iloc[:,:4]) #The df_iris have one column "species" containing non-digit value, 
                                            # so we have to exclude it for svd() to run normally

print('\nThe U'+str(U.shape)+'left-singular vectors matrix of df_iris =\n',U)
print('\nThe S'+str(np.diag(S).shape)+' diagonal sqrt(eigenval) matrix of df_iris =\n',np.diag(S))
print('\nThe Vt'+str(Vt.shape)+' transpose normalized eigenvectors matrix of df_iris =\n',Vt)

#c) Recreate the dataframe from U.dot(np.diag(S)).dot(Vt)
'''
In this example, the U matrix size is (150,150)
meanwhile the S matrix size is only (4,4)

=>Therefore, we can only recreate an approximate matrix of df_iris from U.S.Vt
'''
col_headers = ['sepal_length','sepal_width','petal_length','petal_width']

df_iris_approx = pd.DataFrame(U[:,:4].dot(np.diag(S)).dot(Vt),columns=col_headers)
df_iris_approx.insert(4,'species',df_iris['species'],True) #Add a new column having index = 4 (last), name='species',
                                                           # value of the columns = df_iris['species']
                                                           # True = changes made in the current dataframe

print('\nThe approximate dataframe recreate by U.S.Vt =\n',df_iris_approx.iloc[:10,:])

#d) Calculate the error between df and df_approx
df_error = df_iris.iloc[:,:4] - df_iris_approx.iloc[:,:4]
print('\nThe error between df and df_approx =\n',df_error.iloc[:,:10])

## e) Another way to recreate the df_iris matrix
Sigma = np.zeros_like(df_error) #Create a 0 matrix having the same shape of df_error
                               # so we will have a (150x4) zero matrix

Sigma[:df_error.shape[1], :df_error.shape[1]] = np.diag(S) #The df_error.shape[1] returns the number of column, so it = 4
                    #So this command will fill the first 4x4 zero value of that (150x4) zero matrix with the np.diag(S)
                    # and this Sigma matrix is now literally the np.diag(S) matrix but with more zeros_like rows below

print(Sigma.shape)

df_c = U.dot(Sigma.dot(Vt)) #Recreate the matrix by U.S.Vt
print(df_c[0:5])

df_c_new = pd.DataFrame(df_c, columns = df_error.columns)
print(df_c_new.head())
```
