```python
"""=============================================================================
Ex4: SVD Demensionality Reduction
    Câu 2: 
        a) Tải dữ liệu digits từ dataset của sklearn
        b) Sử dụng TruncatedSVD để giảm chiều dữ liệu xuống 10 components
        c) Trực quan hóa dữ liệu sau khi giảm chiều
https://chrisalbon.com/machine_learning/feature_engineering/dimensionality_reduction_on_sparse_feature_matrix/
============================================================================="""
print('\n-----------------Truncated SVD--------------')
from sklearn.decomposition import TruncatedSVD
from scipy.sparse import csr_matrix
from sklearn import datasets

#a) Load the data name 'digits' from the datasets of sklearn
digits = datasets.load_digits() #This will return a dictionary into variable digits,
                                # to access the data matrix from digits, use digits.data

X = digits.data #Assign the matrix from digits.data into X, 1797x64
print('\nThe matrix X'+str(digits.data.shape)+' containing data of "digits" from datasets of sklearn =\n',X)

X_sparse = csr_matrix(X) #Matrix X has many zero values, so use this to get a more compact version
print('\nThe matrix X_sparse'+str(X_sparse.shape)+' =\n',X_sparse)

#b)Use truncated SVD to compress data into 10 components only
tsvd = TruncatedSVD(n_components=10) #create a truncatedSVD object to do tsvd with only 10 components of input matrix

## Create a 10-dimensional X_sparse_tsvd matrix that retains only 10 principle components of X_sparse (or X)
X_sparse_tsvd = tsvd.fit(X_sparse).transform(X_sparse)
                #the .fit() method will do SVD anlysis on X_sparse matrix to find out 10 most principle components
                #the .transform() method will use the output of .fit() method to create new 10-component matrix

print('\nThe shape of original X or X_sparse matrix =',X.shape)
print('The shape of new 10-dimension X_sparse_tsvd matrx =',X_sparse_tsvd.shape)

print('\nFirst row of original X matrix =\n',X[0])
print('\nFirst row of 10-dimension X_sparse_tsvd matrix =\n',X_sparse_tsvd[0])

## View Percent Of Variance Explained By New Features
# Sum of 10 components' explained variance ratios
print('\nThe sum of variance ratio for retaining 10 components =',tsvd.explained_variance_ratio_[0:10].sum())

## 0.73 = 73% with 10 components
##   it means that, if we retain only 10 components, we will recover only 73% information from the original data
##      and lose 27% information
```
