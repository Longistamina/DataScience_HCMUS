```python
"""=============================================================================
Bài toán 2: Xem xét một bộ dữ liệu 40 mẫu của một nhãn hiệu pin khác (Battery 2).
    Các quan sát được liệt kê trong tập tin life_batteries_2.txt.
   1. Tạo life2_array từ nội dung tập tin.
   2. So sánh 2 nhóm pin ở bài toán 1 và ở bài toán 2 theo histogram và nhận xét
   3. Làm bảng phân phối tần suất cho dữ liệu theo các khoảng:
         120-129, 130-139, 140-149, 150-159, 160 - 169.
   4. Vẽ biểu đồ khối cho bảng phân phối tần suất trên. 
      Biểu đồ câu 2 và câu 4 nói lên điều gì?
   5. So sánh hai nhóm mẫu dựa trên thống kê chung (dùng stats.describe(array))
      Nhận xét kết quả.
   6. Tạo dataframe chứa tần suất của cả 2 nhóm mẫu gồm có 2 cột là batteries_1 
      và batteries_2, index là các khoảng như câu 3
   7. Vẽ boxplot cho cả 2 nhóm batteries_1 và batteries_2 => nhận xét
============================================================================="""
print()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn

from matplotlib.gridspec import GridSpec
from scipy import stats
```

```python
print('==========================================================')
print('*** 1. Read data from .txt files into array***')
print('==========================================================')

f1 = open('data_Buoi7&8/life_batteries.txt')
read_bat1 = f1.read().replace('\n',' ').split()
batt1_life_arr = np.array([ eval(item) for item in read_bat1])
f1.close()

print("\nBatteries' life array_1:\n",batt1_life_arr)

#----------------------

f2 = open('data_Buoi7&8/life_batteries_2.txt')
read_bat2 = f2.read().replace('\n',' ').split()
batt2_life_arr = np.array([ eval(item) for item in read_bat2])
f2.close()

print("\nBatteries' life array_2:\n",batt2_life_arr)
```

```python
print('==========================================================================')
print('*** 2. Compare two groups of batteries life via histogram and comments***')
print('==========================================================================')

fig = plt.figure(figsize=(10,4))
grsp = GridSpec(nrows=1, ncols=2)

batt1_hist = fig.add_subplot(grsp[0,0])
batt1_hist.set_title('Histogram of batt1_life_arr', color='brown')
sbn.histplot(batt1_life_arr, kde=True, color='brown', bins=8)

batt2_hist = fig.add_subplot(grsp[0,1], sharey=batt1_hist)
batt2_hist.set_title('Histogram of batt2_life_arr', color='green')
sbn.histplot(batt2_life_arr, kde=True, color='green')

plt.show()

print('\nComments:')
print('  1: Có nhiều dữ liệu hơn cho Battery 2 trong khoảng 140 so với Battery 1 trong khoảng 130.')
print('  2: Ngoài ra, mức độ biến thiên của Battery 2 ít hơn so với Battery 1.')
print('  3: Dựa trên các kết quả này, có thể kết luận rằng Battery 2 là một nhãn hiệu tốt hơn (trung bình cao hơn và biến thiên thấp hơn).')
print('  4: Tuy nhiên, tính hợp lệ của kết luận này còn phụ thuộc vào cách thu thập dữ liệu.')
```

```python
print('====================================================================================')
print('*** 3. Create distribution frequency table of batt1_life_arr and batt2_life_arr ***')
print('            120-129, 130-139, 140-149, 150-159, 160 - 169             ')
print('=====================================================================================')

batt1_freq = pd.Series()
batt1_freq['120-129'] = np.extract(((batt1_life_arr>=120) & (batt1_life_arr<=129)), batt1_life_arr).size
batt1_freq['130-139'] = np.extract(((batt1_life_arr>=130) & (batt1_life_arr<=139)), batt1_life_arr).size
batt1_freq['140-149'] = np.extract(((batt1_life_arr>=140) & (batt1_life_arr<=149)), batt1_life_arr).size
batt1_freq['150-159'] = np.extract(((batt1_life_arr>=150) & (batt1_life_arr<=159)), batt1_life_arr).size
batt1_freq['160-169'] = np.extract(((batt1_life_arr>=160) & (batt1_life_arr<=169)), batt1_life_arr).size

print('\nFrequency table of batt1_life_arr:\n',batt1_freq)

#-------------------------------------------#
batt2_freq = pd.Series()
batt2_freq['120-129'] = np.extract(((batt2_life_arr>=120) & (batt2_life_arr<=129)), batt2_life_arr).size
batt2_freq['130-139'] = np.extract(((batt2_life_arr>=130) & (batt2_life_arr<=139)), batt2_life_arr).size
batt2_freq['140-149'] = np.extract(((batt2_life_arr>=140) & (batt2_life_arr<=149)), batt2_life_arr).size
batt2_freq['150-159'] = np.extract(((batt2_life_arr>=150) & (batt2_life_arr<=159)), batt2_life_arr).size
batt2_freq['160-169'] = np.extract(((batt2_life_arr>=160) & (batt2_life_arr<=169)), batt2_life_arr).size

print('\nFrequency table of batt2_life_arr:\n',batt2_freq)
```

```python
print('=============================================================================================')
print('*** 4. Draw barplot for distribution frequency table of batt1_life_arr and batt2_life_arr ***')
print('=============================================================================================')
#------visualize
fig = plt.figure(figsize=(10,4))
grsp = GridSpec(nrows=1, ncols=2)

batt1_bar = fig.add_subplot(grsp[0,0])
batt1_bar.set_title('Barplot of Battery_1 life frequency', color='brown')
sbn.barplot(x=batt1_freq.index, y=batt1_freq, color='brown')
batt1_bar.set_xlabel('Battery life (minute)')
batt1_bar.set_ylabel('Count')

batt2_bar = fig.add_subplot(grsp[0,1], sharey=batt1_bar)
batt2_bar.set_title('Barplot of Battery_2 life frequency', color='green')
sbn.barplot(x=batt2_freq.index, y=batt2_freq,color='green')
batt2_bar.set_xlabel('Battery life (minute)')
batt2_bar.set_ylabel('Count')

plt.show()

print('\nComments: The barplot has gaps between intervals, while the histogram does not.')
```

```python
print('==================================================================================')
print('*** 5. Compare batt1_life_arr and batt2_life_arr using scipy.stats.describe()***')
print('==================================================================================')
batt1_desc = stats.describe(batt1_life_arr)
print('\nDescriptive statistics of batt1_life_arr:')
for attr in dir(batt1_desc):
    if attr.startswith('__') or attr.startswith('_'): continue
    elif attr=='index' or attr=='count': continue
    else: print(f'  batt1_life_arr {attr}: {getattr(batt1_desc, attr)}')

#---------------------#
batt2_desc = stats.describe(batt2_life_arr)
print('\nDescriptive statistics of batt2_life_arr:')
for attr in dir(batt2_desc):
    if attr.startswith('__') or attr.startswith('_'): continue
    elif attr=='index' or attr=='count': continue
    else: print(f'  batt2_life_arr {attr}: {getattr(batt2_desc, attr)}')

#---------------------#
print('\nComments:')
print('  1. Battery_2 has higher mean than Battery_1')
print('  2. Battery_2 has lower variance than Battery_1, meaning Battery_2 life is less fluctuated')
print('  3. Battery_2 has excess kurtosis > 0, meaning its kurtosis is higher than normal distribution')
print('  4. Battery_1 has excess kurtosis < 0, meaning its kurtosis is lower than normal distribution')
print('  5. Battery_2 has skewness < 0, meaning its distribution is skew to the left')
print('  6. Battery_1 has skewness > 0, meaning its distribution is skew to the right')
```

```python
print('=============================================================================================')
print('*** 6. Create dataframe for batt1_life_freq and batt2_life_freq, index col has the intervals***')
print('=============================================================================================')

df_batt_freq = pd.DataFrame({'Value_intervals':batt1_freq.index,
                              'Battery_1_frequency':batt1_freq.values,
                             'Battery_2_frequency':batt2_freq.values})

df_batt_freq.set_index('Value_intervals', inplace=True)

df_batt_freq
```

```python
print('=============================================================================================')
print('*** 7. Draw boxplot for Battery_1_life and Battery_2_life, give comments***')
print('=============================================================================================')

df_batt_life = pd.DataFrame({
    'Battery_1_life':batt1_life_arr,
    'Battery_2_life':batt2_life_arr
})

print(df_batt_life.head())
print()
sbn.boxplot(df_batt_life);
```

```python
#--------use pandas.melt() to merge wide-dataframe into long-dataframe----------#
df_batt_life_melt = pd.melt(frame = df_batt_life,
                           var_name='Batteries', #Name for the column that contains variables/columns from original dataframe
                           value_vars=['Battery_1_life','Battery_2_life'], #The variable corresponding to its values from original dataframe
                           value_name='Life(minute)') #Set the name for the column that contains all values from all original dataframe's columns
print(df_batt_life_melt)

print()

sbn.boxplot(data=df_batt_life_melt, x='Batteries', y='Life(minute)', hue='Batteries');
```
