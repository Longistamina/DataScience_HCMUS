```python
"""=============================================================================
Ex4: MATRICES & VECTORS
    Câu 1:
        2a +  b +  c = 4
         a + 3b + 2c = 5
         a           = 6
       
         Quy về Ax = B. Sau đó giải tìm x 
============================================================================"""
import numpy as np
print('----------------CAU 1-------------------------')
mA = np.array([[2,1,1],
               [1,3,2],
               [1,0,0]])

mB = np.array([4,5,6])

print('Matrix A =\n',mA)
print('\nMatrix B =',mB)

print('\nNghiệm của phương trình Ax = B là:',np.linalg.solve(mA,mB),'\n')
```

```python
"""=============================================================================
Ex4: MATRICES & VECTORS
    Câu 2:
        a) Tạo 1 array points có 1000 phần tử cách đều nhau với:
            start = -5, end = 5, step = 0.01
        b) Tạo ma trận điểm có 1000 điểm với x,y = points, point bằng np.meshgrid
        c) Tính z = sqrt(power(x, 2) + power(y, 2))
        d) Biểu diễn z
============================================================================="""
import numpy as np
import matplotlib.pyplot as plt
print('----------------CAU 2-------------------------')

arr_points = np.arange(-5, 5, 0.01) #Create array
print(arr_points[:20])

x,y = np.meshgrid(arr_points, arr_points)
print('\nx[:10] =',x[:10])
print('\ny[:10] =',y[:10])


z = np.sqrt(x**2 + y**2)

plt.figure(figsize=(8,8))
plt.imshow(z, cmap=plt.cm.Blues)
plt.colorbar()
plt.title('Image plot of $\\sqrt{x^2 + y^2}$ for a grid of values')
```
