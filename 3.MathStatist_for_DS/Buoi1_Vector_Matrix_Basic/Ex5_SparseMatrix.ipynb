```python
"""=============================================================================
Ex5: SPARSE MATRIX
    Câu 1:
        a) Tạo ma trận A(5, 5) với các giá trị ngẫu nhiên từ -10 đến 5
        b) Thay thế giá trị 0 cho tất cả các phần tử âm trong ma trận A
        c) Tạo sparse matrix S từ A
        d) Tính sparsity của sparse matrix 
============================================================================="""
print('-------------------CAU 1----------------------------')
import numpy as np
import random
from scipy.sparse import csr_matrix

#---a) Viết hàm để tạo ma trận bất kỳ, tạo ma trận A(5, 5) với các giá trị ngẫu nhiên từ -10 đến 5-------------#
def create_matrix (m,n,start,end):
    mtrx = []
    for i in range(m):
        row = []
        for j in range(n):
            a = random.randint(start,end+1) #Tạo giá trị ngẫu nhiên a trong khoảng start - end+1
            row.append(a) #Thêm giá trị a vào từng cột của dòng hiện hành
        mtrx.append(row) #Thêm dòng mới tạo vào list mtrx
    return np.array(mtrx) #Đổi list mtrx thành matrix thật, rồi return

A = create_matrix(5,5,-10,5)
print('a) Matrix A =\n',A)

#----b) Thay thế giá trị 0 cho tất cả các phần tử âm trong ma trận A
A[A < 0] = 0
print('\nb) Matrix A new =\n',A)

#----c) Tạo sparse matrix S từ A
#Là loại matrix cho biết vị trí của những phần tử khác 0 trong ma trận, và value của nó
print('\nS = CSR =\n', csr_matrix(A))

#----d) Tính sparsity (độ thưa) của sparse matrix
print('\nĐộ thưa của Sparse matrix = %.f' %(100 * (1.0 - (np.count_nonzero(A) / A.size))) + '%')
```

```python
"""=============================================================================
Ex5: SPARSE MATRIX
    Câu 2:
        a) Tạo ra một ma trận BigA(1000, 1000) phần tử ngẫu nhiên tử -10 đến 5.
        b) Thay thế giá trị 0 cho tất cả các phần tử âm trong ma trận BigA
        c) Tính kích thước của ma trận BigA
        d) Tạo sparse matrix BigS từ BigA
        e) Tích kích thước của BigS
        f) Tính sparsity của sparse matrix
        g) Trực quan hóa BigS
============================================================================="""
print('-------------------CAU 2----------------------------')
import numpy as np
import random
import matplotlib.pyplot as plt
from scipy.sparse import csr_matrix

def cre_matrix (m,n,start,end):
    mtr = []
    for i in range(m):
        row = []
        for j in range(n):
            value = random.randint(start,end+1)
            row.append(value)
        mtr.append(row)
    return np.array(mtr)

mtrBigA = cre_matrix(1000,1000,-10,5)
print('a) mtrBigA =\n',mtrBigA)

mtrBigA_0 = np.where(mtrBigA < 0, 0, mtrBigA) #Where mtrBig A is < 0, replace 0 for that value
print('\nb) mtrBigA_0 =\n',mtrBigA_0)

print('\nc) Size of mtrBigA =',mtrBigA.shape[0])

mtrBigS = csr_matrix(mtrBigA_0)
print('\nd) BigS = csrS = Sparse Matrix S\n',csr_matrix(mtrBigA_0))

print('\ne) Size of mtrBigS =',mtrBigS.shape[0])

print('\nf) Sparsity of matrix BigS = %.f' %(100 * (1.0 - (np.count_nonzero(mtrBigA_0) / mtrBigA_0.size))) + '%')

print('\ng) Visualize matrix BigS:')
plt.figure(figsize = (10, 10))
plt.spy(mtrBigS, markersize = 0.1)
```
