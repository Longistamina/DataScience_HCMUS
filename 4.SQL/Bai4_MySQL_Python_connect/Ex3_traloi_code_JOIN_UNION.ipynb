```python
def run_sql(sql):
    # code here
    curs = conn.cursor()
    curs.execute(sql)
    results = curs.fetchall()
    curs.close()
    return results
    
    
```

```python
# Câu 1: Liệt kê danh sách sản phẩm theo từng nhà cung cấp, thông tin gồm ProductId, ProductName, UnitPrice, CompanyName
sql_cmd1 = '''
SELECT p.Id as ProductId, p.ProductName, p.UnitPrice, s.CompanyName
FROM `products` p JOIN `suppliers` s ON p.SupplierId = s.Id;
'''

results_1 = run_sql(sql_cmd1)

for row in results_1:
    print(row)
```

```python
# Câu 2: Liệt danh sách các sản phẩm được đặt hàng bởi 'Thomas Hardy'
# Thông tin gồm: Order_Id, OrderDate, ProductName, Quantity, SalePrice và CustomerName
# Trong đó SalePrice = UnitPrice * Quantity

sql_cmd2 = '''
SELECT o.Id as OrderId, o.OrderDate, p.ProductName, oi.Quantity, 
       (oi.Quantity * oi.UnitPrice) AS SalePrice, 
       CONCAT(c.FirstName, ' ', c.LastName) AS CustomerName
FROM `customers` c JOIN `orders` o ON c.Id = o.CustomerId
                JOIN `orderitems` oi ON o.Id = oi.OrderId 
                JOIN `products` p ON oi.ProductId = p.Id;
'''

results_2 = run_sql(sql_cmd2)

for row in results_2:
    print(row)
```

```python
# Câu 5: Liệt kê danh sách các khách hàng chưa có đơn đặt hàng nào và các nhà cung cấp có sản phẩm ngưng bán
# Thông tin gồm: Type, Id, Name, City, Country, Phone
# Trong đó: 
#          + Type: cho biết đó là Customer hay Supplier
#          + Id: hiển thị Id của Customer hoặc Id của Supplier tùy theo Type
#          + Name: hiển thị CustomerName hoặc CompanyName tùy theo Type

sql_cmd5 = '''
(SELECT 'Customer' as Type, 
 c.Id, CONCAT(c.FirstName, ' ', c.LastName) as Name, c.City, c.Country, c.Phone
FROM `customers` c  LEFT JOIN `orders` o ON c.Id = o.CustomerId
WHERE o.CustomerId IS NULL)
UNION
(SELECT 'Supplier' as Type, 
 s.Id, s.CompanyName as Name, s.City, s.Country, s.Phone
FROM `products` p JOIN `suppliers` s ON p.SupplierId = s.Id
WHERE p.IsDiscontinued=1);
'''

results_5 = run_sql(sql_cmd5)

for row in results_5:
    print(row)
```

```python
# Câu 6: Liệt kê thông tin đặt hàng của các khách hàng (kể cả các khách hàng chưa có đặt hàng), chỉ liệt kê 10 dòng đầu tiên.
sql_cmd6 = '''
SELECT CONCAT(c.FirstName, ' ', c.LastName) AS NAME,
	   o.Id as OrderId, o.OrderDate
FROM `orders` o RIGHT JOIN `customers` c ON o.CustomerId=c.Id
ORDER BY o.Id
LIMIT 10;
'''

results_6 = run_sql(sql_cmd6)

for row in results_6:
    print(row)
```
