# Câu 1: Cho dữ liệu employees1.csv và employees2.csv 
## Đọc dữ liệu từ 2 tập tin trên vào 2 DataFrame với index của các DataFrame là 'employee'
## Kết hợp 2 DataFrame trên thành 1 DataFrame dùng chung 1 index là 'employee' (gợi ý: dùng left_index hoặc/và right_index hoặc dùng dataframe1.join(dataframe2))

```python
import numpy as np
import pandas as pd
```

```python
employees1 = pd.read_csv("data3/Ex4_Combining_Data/employees1.csv", index_col='employee')
employees1.drop('Unnamed: 0', axis=1, inplace=True)
employees1
```

```python
employees2 = pd.read_csv("data3/Ex4_Combining_Data/employees2.csv", index_col='employee')
employees2.drop('Unnamed: 0', axis=1, inplace=True)
employees2
```

```python
employees = pd.merge(employees1, employees2, left_index=True, right_index=True)
#left_index=True to use index column of left table as joint key
#right_index=True to use index column of right table as joint key

employees
```

# Câu 2: Cho dữ liệu department.csv 
## Đọc dữ liệu từ tập tin trên 
## Kết hợp dữ liệu này với dữ liệu kết quả từ câu 1 

```python
department = pd.read_csv('data3/Ex4_Combining_Data/department.csv', index_col=0)
department
```

```python
emp_dep = employees.merge(department, on='group')
emp_dep.insert(0, 'employee', employees.index)
emp_dep
```

# Câu 3: Cho dữ liệu skills.csv 
## Đọc dữ liệu từ tập tin trên 
## Kết hợp dữ liệu này với dữ liệu kết quả từ câu 2

```python
skills = pd.read_csv("data3/Ex4_Combining_Data/skills.csv", index_col=0)
skills
```

```python
emp_dep_skill = emp_dep.merge(skills, on='group')

skills_col = emp_dep_skill.pop('skills')
emp_dep_skill.insert(2, 'skills', skills_col)

emp_dep_skill
```

# Câu 4: Cho dữ liệu salary.csv 
## Đọc dữ liệu từ tập tin trên 
## Kết hợp dữ liệu này với dữ liệu ở câu 1 (gợi ý: dùng right_on & left_on khi merge vì trong salay có cột name, còn df ở câu 1 lại có cột employee trùng nội dung, bỏ cột name) 

```python
salary = pd.read_csv("data3/Ex4_Combining_Data/salary.csv", index_col=0)
salary
```

```python
emp_dep_skill_salary = pd.merge(emp_dep_skill, salary, left_on="employee", right_on='name')
#The 'employee' col of left table and 'name' col of right table share the same value, but the col names are different
#Therefore use left_on and right_on to define these cols that differ in name but share the same values row-by-row as joint key

emp_dep_skill_salary
```
