## Đọc dữ liệu 

```python
billboard = pd.read_csv("data3/Ex2_Tidying_Data/billboard.csv", sep=',', encoding="mac_latin2")
billboard
```

## Xem xét vấn đề về dữ liệu cần khắc phục 

```python
print("Comment: The 'x...week' should be the values of a column, not column names")
```

## Chuẩn lại dữ liệu để khắc phục vấn dề trên 

```python
billboard_melt = pd.melt(billboard, id_vars=['year', 'artist.inverted', 'track', 'time', 'genre',	'date.entered',	'date.peaked'], var_name='week', value_name='rank')
billboard_melt
```

## Sau khi chuẩn lại dữ liệu, có điều gì phát sinh cần khắc phục tiếp theo không? Nếu có thì hãy chuẩn lại dữ liệu mới có ở câu trên

```python
print("Comment: Many values in the rank column are NaN\n")

billboard_melt.info()
```

```python
print("Drop all NaN values:\n")
billboard_melt=billboard_melt.dropna()
billboard_melt.info()
```

```python
#Change the values in "week" column from string to numeric, for example from x1stweek => 1
#The "weeK" column dtype is object, meaning it contains string value
billboard_melt['week'] = billboard_melt['week'].str.extract('(\d+)')
billboard_melt['week'] = billboard_melt['week'].astype('int')

import warnings
warnings.filterwarnings('ignore')

billboard_melt
```

```python
billboard_melt.info()
```

## Vấn đề  mới phát sinh: có nhiều giá trị trong cột "artist.inverted" và cột "rank" bị lặp lại
## => Do đó, cần tách ra thành hai dataframe nhỏ hơn

```python
#create df_songs

df_songs = billboard_melt[['year', 'artist.inverted',	'track', 'time', 'genre']]
df_songs=df_songs.drop_duplicates() #To drop duplicate values
df_songs=df_songs.reset_index(drop=True)
df_songs['song_id']=df_songs.index #Add song_id as row indexes of the dataframe
df_songs
```

```python
#create df_ranks

df_ranks = pd.merge(billboard_melt, df_songs, on=['year', 'artist.inverted', 'track', 'time', 'genre'])
df_ranks
```

```python
#remove "track" column as it is identified by the song_id now
df_ranks = df_ranks.drop('track', axis=1)
df_ranks
```

```python
#Add "date" column to convert all date_time into year 2000 period
df_ranks['date'] = pd.to_datetime(df_ranks['date.entered']) + pd.to_timedelta(df_ranks['week'], unit='W')
df_ranks = df_ranks.sort_values(by=['song_id','week'], ascending=True) #Sort the df_ranks to see the changes in date

df_ranks = df_ranks[['song_id','date','week','rank']] #Retain only these columns
df_ranks
```
