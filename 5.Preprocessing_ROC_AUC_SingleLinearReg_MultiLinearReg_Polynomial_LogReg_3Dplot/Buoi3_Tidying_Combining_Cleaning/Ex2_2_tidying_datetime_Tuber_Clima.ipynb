```python
import numpy as np
import pandas as pd
```

```python
tuber = pd.read_csv("data3/Ex2_Tidying_Data/tb-raw.csv")
tuber
```

```python
print('Problems:\n')
print('### Age range and sex are stored in the same cell: f014 is female aged 0-14')
print('### Age range and sex must be values, not column names')
```

```python
tuber_melt = pd.melt(tuber, id_vars=['country','year'], var_name='sex_age', value_name='cases')
tuber_melt = tuber_melt.dropna()
tuber_melt
```

```python
tuber_melt[['sex','lower_age','upper_age']] = tuber_melt['sex_age'].str.extract('(m|f)(\d+)(\d{2})', expand=True)
# expand=True to pass the extracted string into adjacent columns
#This will extract the string that matchs (m|f) pattern and pass into 'sex' column
#          extract the string that matchs (\d+) pattern and pass into 'lower_age' column
#          extract the string that matchs (\d{2}) pattern and pass into 'upper_age' column
tuber_melt
```

```python
tuber_melt['age'] = tuber_melt['lower_age'] + "-" + tuber_melt['upper_age'] #Create age group string: 0-14, 15-24,...
tuber_melt = tuber_melt.drop(['sex_age','lower_age','upper_age'], axis=1).reset_index(drop=True) #remove columns that no longer need

#bring the column 'cases' to the last of dataframe
case_count = tuber_melt.pop('cases') #Pop the column "cases" out, store in column variable
tuber_melt.insert(tuber_melt.shape[1], column='cases', value=case_count) #Insert the column "cases" back in the position -1 (last of dataframe)

tuber_melt
```

# Câu 4: Global Historical Climatology Network 
## Cho dữ liệu weather-raw.csv. Bộ dữ liệu này đại diện cho các bản ghi thời tiết hàng ngày cho một trạm thời tiết (MX17004) ở Mexico trong 5 tháng năm 2010. 
## Đọc dữ liệu 
## Xem xét vấn đề về dữ liệu cần khắc phục 
## Chuẩn lại dữ liệu để khắc phục vấn đề trên

```python
weather = pd.read_csv("data3/Ex2_Tidying_Data/weather-raw.csv")
weather
```

```python
weather_melt = pd.melt(weather, id_vars=['id','year','month','element'] , var_name='day', value_name='record_value')
weather_melt
```

```python
#Convert d1 => 1, d2 => 2, ....
weather_melt['day'] = weather_melt['day'].str.extract('(\d{1})', expand=False)
weather_melt['day'] = weather_melt['day'].astype(int)
weather_melt
```

```python
weather_melt = weather_melt.dropna().reset_index(drop=True)
weather_melt['id'] = weather_melt['id'].str.strip()
weather_melt
```

## Split tmax and tmin into two separate dataframes, then merge them into new one

```python
weather_tempmax = weather_melt[weather_melt.element == 'tmax'][['id','year','month','day','record_value']]
weather_tempmax.columns = ['id','year','month','day','temp_max']
weather_tempmax
```

```python
weather_tempmin = weather_melt[weather_melt.element == 'tmin'][['id','year','month','day','record_value']]
weather_tempmin.columns = ['id','year','month','day','temp_min']
weather_tempmin
```

```python
weather_final = pd.merge(weather_tempmax, weather_tempmin, on=['id','year','month','day'])
weather_final
```

```python
from datetime import datetime
#Create datetime string from 'year', 'month' and 'day' columns
def create_date_string (df):
    date_string = datetime(year=df['year'], month=df['month'], day=df['day'])
    return date_string

weather_final.insert(1, 'date', weather_final.apply(create_date_string, axis=1))

weather_final.drop(['year','month','day'], axis=1, inplace=True)

weather_final
```
