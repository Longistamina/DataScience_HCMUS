# Câu 1: 
## Cho dữ liệu BMI.xlsx. Bộ dữ liệu này ghi lại BMI của một số quốc gia trong một số năm. 

## 1_Đọc dữ liệu
## 2_Xem xét vấn đề về dữ liệu cần khắc phục 
## 3_Chuẩn lại dữ liệu để khắc phục vấn dề trên 
## 4_Đặt yêu cầu ngược lại là cần phải tạo pivot table để xem thống kê theo country, year. 
## Hãy chuyển dữ liệu mới làm ở trên về dạng thống kê

```python
import numpy as np
import pandas as pd
```

## 1_Đọc dữ liệu

```python
bmi = pd.read_excel("data3/Ex1_Tydying_Data/BMI.xlsx")
bmi
```

## 2_Xem xét vấn đề về dữ liệu cần khắc phục 

```python
print("Comment: The years Y1980, Y1981... should be values, not column names")
```

## 3_Chuẩn lại dữ liệu để khắc phục vấn dề trên

```python
bmi_melt = pd.melt(bmi, id_vars="Country", value_vars=["Y1980", "Y1981","Y1982","Y1983"], var_name="Year", value_name="BMI")
bmi_melt
```

## 4_Đặt yêu cầu ngược lại là cần phải tạo pivot table để xem thống kê theo country, year. 
## Hãy chuyển dữ liệu mới làm ở trên về dạng thống kê

```python
bmi_melt_pivot = pd.pivot(bmi_melt,
                         index="Country",
                         columns="Year",
                         values="BMI")
bmi_melt_pivot
```

# Câu 2: 
## Cho dữ liệu student.xlsx. Bộ dữ liệu này ghi lại điểm các môn của sinh viên 
## 1_Đọc dữ liệu 
## 2_Xem xét vấn đề về dữ liệu cần khắc phục 
## 3_Chuẩn lại dữ liệu để khắc phục vấn dề trên 
## 4_Đặt yêu cầu ngược lại là cần phải tạo pivot table để xem thống kê theo từng student và các subject. 
## Hãy chuyển dữ liệu mới làm ở trên về dạng thống kê

## 1_Đọc dữ liệu 

```python
student = pd.read_excel("data3/Ex1_Tydying_Data/student.xlsx")
student
```

## 2_Xem xét vấn đề về dữ liệu cần khắc phục 

```python
print("The subjects' names should be values of a column, not stand as column names")
```

## 3_Chuẩn lại dữ liệu để khắc phục vấn dề trên 

```python
student_melt = pd.melt(student, id_vars="Student", var_name="Subject", value_name="Score")
#If not define value_vars, pd.melt() will under stand all other columns apart from id_vars are value_vars

student_melt
```

## 4_Đặt yêu cầu ngược lại là cần phải tạo pivot table để xem thống kê theo từng student và các subject. 
## Hãy chuyển dữ liệu mới làm ở trên về dạng thống kê

```python
student_melt_pivot = pd.pivot(student_melt,
                             index="Student",
                             columns="Subject")
student_melt_pivot
```
