## Câu 2: billboard.csv

```python
df2 = pd.read_csv('data/billboard.csv', encoding='mac_latin2')
```

```python
df2.head()
```

#### Vấn đề cần khắc phục:

* Tiêu đề cột chứa giá trị thay vì chứa biến
* Trong từng cell chứa thứ hạng (rank)
* Nếu một bài hát nằm trong bảng xếp hạng dưới 75 tuần --> các cột còn lại đếu chứa giá trị null (NaN)

```python
df2_new = pd.melt(frame=df2, id_vars=['year', 'artist.inverted', 'track', 'time', 
                                      'genre', 'date.entered', 'date.peaked'],
                 var_name='week', value_name='rank')
```

```python
df2_new.head(10)
```

```python
df2_new.tail()
```

```python
df2_new.info()
```

#### Cần chuẩn hóa dữ liệu

* Thay giá trị của week từ chuỗi thành số
* Xóa bỏ các dòng mà cột rank có giá trị NaN
* Chuyển kiểu dữ liệu của rank thành kiểu int

```python
df2_new['week'] = df2_new['week'].str.extract('(\d+)', expand=False)
df2_new['week'] = df2_new['week'].astype(int)
```

```python
df2_new.head()
```

```python
df2_new = df2_new.dropna()
```

```python
df2_new['rank'] = df2_new['rank'].astype(int)
```

```python
df2_new.info()
```

```python
df2_new.head()
```

#### Vấn đề mới phát sinh:

* Nhiều đơn vị quan sát (bao gồm song và rank) trong cùng một bảng

```python
# tách các thông tin liên quan đến bài hát --> lưu vào dataframe songs
# các thông tin liên quan đến xếp hạng --> lưu vào dataframe ranks
```

```python
df_songs = df2_new[['year', 'artist.inverted', 'track', 'time', 'genre']]
df_songs = df_songs.drop_duplicates()
df_songs.head(10)
```

```python
df_songs = df_songs.reset_index(drop=True)
df_songs.head()
```

```python
df_songs['song_id'] = df_songs.index
df_songs.head()
```

```python
# tạo dataframe ranks
df_ranks = pd.merge(df2_new, df_songs, 
                    on=['year', 'artist.inverted', 'track', 'time', 'genre'])
df_ranks.head()
```

```python
df_ranks = df_ranks[['song_id', 'date.entered', 'date.peaked', 'week', 'rank']]
df_ranks.head()
```

```python
# tạo cột date
df_ranks['date'] = pd.to_datetime(df_ranks['date.entered']) + \
                    pd.to_timedelta(df_ranks['week'], unit='w') - pd.DateOffset(weeks=1)
df_ranks.head()
```

```python
df_ranks = df_ranks[['song_id', 'date', 'week', 'rank']]
df_ranks.head()
```

## Câu 3: tb-raw.csv

```python
df3 = pd.read_csv('data/tb-raw.csv')
```

```python
df3
```

#### Vấn đề cần khắc phục:

* Tên cột chứa giá trị thay vì chứa biến
* Trong cell chứa giá trị hỗn hợp gồm value, 0, NaN

```python
df3_new = pd.melt(frame=df3, id_vars=['country', 'year'], 
                  var_name='sex_and_age', value_name='cases')
df3_new.head()
```

```python
df3_new.tail()
```

```python
df3_new = df3_new.dropna()
df3_new.tail()
```

#### Vấn đề phát sinh:

* Trong cùng một cột chứa 2 thông tin là Sex và Age

```python
df_temp = df3_new['sex_and_age'].str.extract('(\D)(\d+)(\d{2})')
df_temp.columns=['sex', 'age_lower', 'age_upper']
df_temp['age'] = df_temp['age_lower'] + "-" + df_temp['age_upper']
df_temp.head()
```

```python
df3_new = pd.concat([df3_new, df_temp], axis=1)
df3_new.head()
```

```python
df3_new = df3_new[['country', 'year', 'sex', 'age', 'cases']]
df3_new.head()
```

## Câu 4: weather-raw.csv

```python
df4 = pd.read_csv('data/weather-raw.csv')
```

```python
df4
```

#### Vấn đề cần khắc phục:

* Tên cột chứa giá trị (ngày trong tuần) thay vì chứa biến
* Trong cell chứa giá trị temperature nhưng có rất nhiều giá trị NaN
* biến được lưu trữ trong cả dòng và cột

```python
df4_new = pd.melt(frame=df4, id_vars=['id', 'year', 'month', 'element'],
                 var_name='day', value_name='temp')
df4_new.head()
```

```python
df4_new = df4_new.dropna()
df4_new['id'] = df4_new['id'].str.strip()
df4_new
```

```python
df_max = df4_new[df4_new['element']=='tmax']
df_max.columns=['id', 'year', 'month', 'element', 'day', 'temp_max']
df_max = df_max.drop('element', axis=1)
df_max
```

```python
df_min = df4_new[df4_new['element']=='tmin']
df_min.columns=['id', 'year', 'month', 'element', 'day', 'temp_min']
df_min = df_min.drop('element', axis=1)
df_min
```

```python
df4_final = pd.merge(df_max, df_min, on=['id', 'year', 'month', 'day'])
df4_final
```

```python
df4_final['day'] = df4_final['day'].str.replace('d','')
```

```python
df4_final
```

```python
df4_final[['year', 'month', 'day']] = df4_final[['year', 'month', 'day']].astype(int)
```

```python
import datetime as dt
def create_date(row):
    ngay = dt.datetime(year=row['year'], month=row['month'], day=row['day'])
    return ngay
```

```python
df4_final['date'] = df4_final.apply(create_date, axis=1)
```

```python
df4_final
```

```python
df4_final = df4_final[['id', 'date', 'temp_max', 'temp_min']]
df4_final
```
