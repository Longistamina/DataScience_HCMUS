## Illinois babies' names are the same data, but stored in different .csv files
## Let's merge them

```python
#Tryout glob command
lst_file_path = gl.glob("data3/Ex3_Combining_Data/201*-baby-names-illinois.csv") #glob all target files, store the paths to file_name
print("List of file_path:\n",lst_file_path)

print()
print("File name:")
for file_path in lst_file_path:
    file_name = os.path.basename(file_path)
    print(f'{file_name} .............. shape {pd.read_csv(file_path).shape}')
```

```python
#Read dataframes into lst_data
import re

lst_data=[]
for file_path in lst_file_path:
    data = pd.read_csv(file_path, index_col=None, header=0)
    year = re.findall('(201.)',file_path)[0] #Find the year in file_path using re
    data['year'] = year
    lst_data.append(data)

lst_data
```

```python
illinois_baby_name = pd.concat(lst_data, axis=0)
illinois_baby_name
```
