```python
# Câu 2: Tạo pie chart thể hiện tỷ lệ hành khách nam/nữ trên tàu.
'''
Gợi ý:
- Tạo biến males, female là tổng nam và tổng nữ. 
- Tạo biến proportions là list có 2 phần tử là male và female
- Vẽ biểu đồ: với dữ liệu là proportions, nhãn là ['Males', 'Female'], màu là ['blue', 'red']
- Thiết lập title là Sex Proportion
'''

# Tạo biến males, female là tổng nam và tổng nữ
males = titanic[titanic.Sex =='male'].shape[0] #Lọc ra những giá trị có cột Sex là male, rồi .shape[0] để  lấy tổng số hàng thoả điều kiện
females = titanic[titanic.Sex=='female'].shape[0]

# Tạo biến proportions là list có 2 phần tử là male và female
proportions = [males,females]

# Vẽ biểu đồ
plt.pie(proportions, labels=['Male','Female'], autopct='%.0f%%', explode=(0,0.1), #explode = (0,0.1) để hai phần rẻ quạt tách nhau ra
       colors = ['blue','red']) 

# Thiết lập title là Sex Proportion
plt.title('Sex Proportion of Titanic', color = 'red')

# Show the plot
plt.show()
```

```python
# Câu 4: Vẽ biểu đồ histogram của cột vé (Fare)
'''
Gợi ý:
- Sắp xếp lại dữ liệu df theo cột Fare giảm dần. In head.
- Tạo binsVal = np.arange(0,600,10)
- Vẽ histogram với dữ liệu là df, bins = binsVal. 
- Bổ sung các thông tin xlabel, ylabel, title
'''
# Sắp xếp lại dữ liệu df theo cột Fare giảm dần, và hiển thị các dòng dữ liệu đầu của df
fare = sorted(titanic.Fare, reverse=True)

binsVal = np.arange(0,600,10) #Chia dữ liệu thành 10 khoảng tương ứng với 10 cột, cận dưới là 0, cận trên là 600

plt.hist(fare, bins=binsVal,edgecolor='brown')

plt.xlabel('Giá vé (Bảng Anh)',size=15)
plt.ylabel('Tần số',size=15)
plt.title('HISTOGRAM CỦA GIÁ VÉ TÀU TITANIC', size=20,color='r')
plt.show()

# Bạn nhận xét gì về biểu đồ vừa vẽ => Đa số vé được mua có giá từ 50 Bảng Anh đổ lại, giá vé đắt tiền hạng sang thì ít hơn
```
