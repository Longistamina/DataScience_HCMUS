```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

#GridSpec cũng dùng để  vẽ nhiều biểu đồ trong một figure giống subplot, 
# nhưng gridspec sẽ tiện hơn trong việc sửa chữa và merge các object đồ thị lại với nhau
#  đổi lại thì sẽ phải khai báo từng đối tượng cho gridspect
```

```python
#Tạo dữ liệu
men_means = [20, 35, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
x = np.arange(len(men_means))
w = 0.3 # width
```

```python
fig = plt.figure(figsize=(7,4));

grid_spec = gridspec.GridSpec(nrows=1, ncols=2) #Tạo gridspec gồm 1 dòng và hai cột, tức vẽ được hai đồ thị
axes_1 = fig.add_subplot(grid_spec[0,0]) #Khai báo cho object đồ thị axes_1
axes_2 = fig.add_subplot(grid_spec[0,1]) #Khai báo cho object đồ thị axes_2

axes_1.bar(x, men_means, color='r', align='center', label='Men') #Vẽ đồ thị barplot cho object đồ thị axes_1
axes_2.bar(x, women_means, color='g', align='center', label='Women') #Vẽ đồ thị barplot cho object đồ thị axes_2
axes_1.legend()
axes_2.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(10,7));
grid_spec = gridspec.GridSpec(nrows=2, ncols=2) #Tạo grid spec 2 dòng 2 cột, vẽ được tối đa 4 đồ thị

axes_1 = fig.add_subplot(grid_spec[0,0]) #Khai báo object đồ thị axes_1 ở vị dòng 1 cột 1
axes_2 = fig.add_subplot(grid_spec[0,1]) #Khai báo object đồ thị axes_2 ở vị trí dòng 1 cột 2
axes_3 = fig.add_subplot(grid_spec[1,:]) #Khai báo object đồ thị axes_3 ở dòng 2, và chiếm luôn cả hai cột (nên vẽ ra thấy khung dài hơn)

axes_1.bar(x, men_means, color='r', align='center', label='Men')
axes_2.bar(x, women_means, color='g', align='center', label='Women')
axes_3.bar(x, men_means, color='r', align='center', width=w, label='Men')
axes_3.bar(x+w, women_means, color='g', align='center', width=w, label='Women')

axes_1.set_xlabel('Scores by group of Men')
axes_1.set_ylabel('Scores')
axes_2.set_xlabel('Scores by group of Women')
axes_2.set_ylabel('Scores')
axes_3.set_xlabel('Scores by group of Men and Women')
axes_3.set_ylabel('Scores')
axes_3.legend()
fig.suptitle('Scores by group')

plt.show()
```

```python
fig = plt.figure(figsize=(10,7));
grid_spec = gridspec.GridSpec(nrows=2, ncols=2) #Tạo grid spec 2 dòng 2 cột, vẽ được tối đa 4 đồ thị

axes_1 = fig.add_subplot(grid_spec[0,0]) #Khai báo object đồ thị axes_1 ở vị dòng 1 cột 1
axes_2 = fig.add_subplot(grid_spec[1,0]) #Khai báo object đồ thị axes_2 ở vị trí dòng 2 cột 1
axes_3 = fig.add_subplot(grid_spec[:,1]) #Khai báo object đồ thị axes_3 ở cột 2, và chiếm luôn cả hai dòng (vẽ ra thấy khung cao hơn)

axes_1.bar(x, men_means, color='r', align='center', label='Men')
axes_2.bar(x, women_means, color='g', align='center', label='Women')
axes_3.bar(x, men_means, color='r', align='center', width=w, label='Men')
axes_3.bar(x+w, women_means, color='g', align='center', width=w, label='Women')

axes_1.set_xlabel('Scores by group of Men')
axes_1.set_ylabel('Scores')
axes_2.set_xlabel('Scores by group of Women')
axes_2.set_ylabel('Scores')
axes_3.set_xlabel('Scores by group of Men and Women')
axes_3.set_ylabel('Scores')
axes_3.legend()
fig.suptitle('Scores by group')

plt.show()
```
