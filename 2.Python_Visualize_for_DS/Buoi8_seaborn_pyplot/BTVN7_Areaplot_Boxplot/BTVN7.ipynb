```python
# Vẽ boxplot cho dữ liệu height và weight
fig = plt.figure(figsize=(8,6))

grid_spec = grsp.GridSpec(nrows=1, ncols=2) #Tạo gridspec gồm 1 dòng và 2 cột (tổng cộng 2 ô đồ thị)
box_height = fig.add_subplot(grid_spec[0,0])
box_weight= fig.add_subplot(grid_spec[0,1])

box_height.boxplot(df_bsbl.height,notch=True, boxprops=dict(color='#8e8883'))
box_height.set_xlabel("Boxplot of baseball players' height")
box_height.set_ylabel('Height (meter)')

box_weight.boxplot(df_bsbl.weight,notch=True, boxprops=dict(color='red'))
box_weight.set_xlabel("Boxplot of baseball players' weight")
box_weight.set_ylabel('Weight (kg)')

fig.tight_layout()
plt.legend
plt.show()
```

#### Kiểm tra xem dữ liệu có outliers hay không? Nếu có thì loại bỏ các outliers. Vẽ lại boxplot

```python
#Tạo hàm để  tìm outliers
def find_outliers(dataset): 
    q1 = dataset.quantile(0.25)
    q3 = dataset.quantile(0.75)
    IQR = q3-q1
    outliers = dataset[ (dataset < (q1-1.5*IQR)) | (dataset > (q3+1.5*IQR))]
    return outliers
```

```python
#Tìm outliers cho height bằng hàm vừa tạo
out_height = find_outliers(df_bsbl.height)
out_height
```

```python
#Tìm outliers cho weight bằng hàm vừa tạo
out_weight = find_outliers(df_bsbl.weight)
out_weight
```

```python
# Loại bỏ các outliers
height_new = df_bsbl.height.to_list()
weight_new = df_bsbl.weight.to_list()

height_new.remove(list(out_height))
weight_new.remove(list(out_weight))
```

```python
# Vẽ lại box plot

fig = plt.figure(figsize=(8,6))

grid_spec = grsp.GridSpec(nrows=1, ncols=2) #Tạo gridspec gồm 1 dòng và 2 cột (tổng cộng 2 ô đồ thị)
box_height = fig.add_subplot(grid_spec[0,0])
box_weight= fig.add_subplot(grid_spec[0,1])

box_height.boxplot(height_new,notch=True, boxprops=dict(color='#8e8883'))
box_height.set_xlabel("Boxplot of baseball players' height")
box_height.set_ylabel('Height (meter)')

box_weight.boxplot(weight_new,notch=True, boxprops=dict(color='red'))
box_weight.set_xlabel("Boxplot of baseball players' weight")
box_weight.set_ylabel('Weight (kg)')

fig.tight_layout()
plt.legend
plt.show()
```
