```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import squarify
```

```python
# with 2 lists
size_lst = [7,18,20,25]
label_lst = ['Excellent', 'Very Good', 'Good', 'Average']
```

```python
# ve tree map
plt.figure(figsize=(8,6))
squarify.plot(sizes=size_lst, label=label_lst)
plt.axis('off')
plt.show()
```

```python
# ve tree map cho 5 lop 10A1, ..., 10A5 voi so hoc sinh cua moi lop
cac_lop = ['10A1', '10A2', '10A3', '10A4', '10A5']
si_so = [45, 50, 47, 52, 46]
```

```python
plt.figure(figsize=(8,6))
squarify.plot(sizes=si_so, label=cac_lop)
plt.axis('off')
plt.show()
```

```python
cac_lop = ['10A1', '10A2', '10A3', '10A4', '10A5']
si_so = [45, 50, 47, 52, 46]
df = pd.DataFrame({'lop':cac_lop, 'so_hoc_sinh':si_so})
```

```python
df
```

```python
plt.figure(figsize=(8,6))
squarify.plot(sizes=df['so_hoc_sinh'], label=df['lop'])
plt.axis('off')
plt.show()
```

```python
labels = df.apply(lambda x: x[0]+'\n('+str(x[1])+')', axis=1)
colors = [plt.cm.Spectral(i/float(len(labels))) for i in range(len(labels))] # phát sinh màu ngẫu nhiên RGB
```

```python
plt.figure(figsize=(8,6))
squarify.plot(sizes=df['so_hoc_sinh'], label=labels, color=colors, alpha=.6)

# Decorate
plt.title('Si so cac lop 10')
plt.axis('off')
plt.show()
```
