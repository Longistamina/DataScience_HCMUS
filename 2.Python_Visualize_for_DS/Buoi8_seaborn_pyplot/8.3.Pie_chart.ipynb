```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb #Nhớ cài pip install seaborn trước trong terminal rồi mới xài
```

```python
df = pd.DataFrame({'lop':['10A1','10A2','10A3','10A4'],
                   'siso':[45,50,36,27]})
df
```

```python
plt.pie(x=df.siso, labels=df.lop) #labels là đặt tên cho từng phần của đồ thị hình tròn


plt.show()
```

```python
plt.pie(x=df.siso, labels=df.lop, #labels là đặt tên cho từng phần của đồ thị hình tròn
       counterclock=False,         #counterclock=False thì sẽ quay đúng chiều kim đồng hồ
       shadow=True,                #shadow=True tạo hiệu ứng bóng đổ  3 chiều
       autopct='%.0f%%')           #Format cho hiện ra phần trăm diện tích mà vùng đó chiếm trong hình tròn

plt.legend()
plt.show()
```
