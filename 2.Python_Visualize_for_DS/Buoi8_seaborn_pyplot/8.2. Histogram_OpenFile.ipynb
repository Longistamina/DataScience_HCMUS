```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb #Nhớ cài pip install seaborn trước trong terminal rồi mới xài
```

```python
#Mở file heights_1.txt ra để đọc dữ liệu, rồi đóng lại
#Dùng hàm eval để đọc dữ liệu, nó tự đánh giá chuỗi đó, nếu thấy toàn là số thì chuyển kiểu dữ liệu từ string sang numeric giùm luôn

f = open('heights_1.txt')
heights = eval(f.read())
f.close()
```

```python
heights = heights
```

```python
type(heights) #Kiểm tra kiểu dữ liệu của heights, đang ở dạng list
```

```python
heights = pd.Series(heights, dtype=np.float32)*0.0254  #Đổi sang dạng series để làm việc được với numpy và pandas
                                                       #Nhân với 0.0254 để  đổi đơn vị từ inch sang mét
                                                       #Chỉ sau khi đổi sang series thì mới nhân vào như vậy được
```

```python
heights.head()
```

```python
plt.hist(heights, color='#99EDC3', edgecolor='green') #Vẽ đồ thị tần số histogram cho biến heights bằng pyplot

plt.xlabel('Chiều cao (m)')
plt.ylabel('Tần số')
plt.title('BIỂU ĐỒ HISTOGRAM CỦA CHIỀU CAO CÁC CẦU THỦ BÓNG RỔ')

plt.show()
```
