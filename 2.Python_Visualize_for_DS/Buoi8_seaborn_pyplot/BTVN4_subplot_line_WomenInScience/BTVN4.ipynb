```python
# Câu 3: Vẽ 2 biểu đồ line plot ở câu 1 nhưng trên 2 subplot: 
# subplot 1: plt.subplot(1, 2, 1), và subplot 2: plt.subplot(1, 2, 2), 
# Lưu ý: sử dụng plt.tight_layout() trước khi show()

# Vẽ biểu đồ line plot cho subplot 1
plt.subplot(1,2,1) #Tạo khu vực gồm 1 hàng và hai cột => 2 ô => đồ thị này sẽ nằm ở ô thứ nhất, bên trái
plt.plot(year,physical_sciences,color='blue')
plt.title('%Women in Physics') #Phải đặt tên riêng cho từng subplot, cả legend cũng vậy

# Vẽ biểu đồ line plot cho subplot 2
plt.subplot(1,2,2) #đồ thị này sẽ nằm ở ô thứ nhất, bên phải
plt.plot(year,computer_science,color='red')
plt.title('%Women in Computer Sciences')

# Sử dụng plt.tight_layout() trước khi show()

plt.tight_layout() #Để căn chỉnh các đồ thị sao cho chúng không bị overlap với nhau
plt.show()
```

```python
# Câu 4: Vẽ 4 biểu đồ line với 4 màu khác nhau: 
# Biểu đồ 1: thể hiện tỷ lệ % bằng Khoa học vật lý được trao cho phụ nữ qua các năm (year - physical_sciences) 
# Biểu đồ 2: thể hiện tỷ lệ % bằng Khoa học máy tính được trao cho phụ nữ qua các năm (year - computer_science)
# Biểu đồ 2: thể hiện tỷ lệ %  phụ nữ tham gia các công việc liên quan đến y tế qua các năm (year - health)
# Biểu đồ 4: thể hiện tỷ lệ %  phụ nữ tham gia các công việc liên quan đến giáo dục qua các năm (year - education)

# trên 4 subplot: plt.subplot(2, 2, 1), plt.subplot(2, 2, 2), plt.subplot(2, 2, 3), plt.subplot(2, 2, 4)

# Vẽ biểu đồ line cho line 1
plt.subplot(2, 2, 1)
plt.plot(year,physical_sciences,color='blue')
plt.title('%Women in Physics')

# Vẽ biểu đồ line cho line 2 
plt.subplot(2, 2, 2)
plt.plot(year,computer_science,color='red')
plt.title('%Women in Computer Sciences')

# Vẽ biểu đồ line cho line 3
plt.subplot(2, 2, 3)
plt.plot(year,health,color='green')
plt.title('%Women in Health field')

# Vẽ biểu đồ line cho line 4
plt.subplot(2, 2, 4)
plt.plot(year,education,color='yellow')
plt.title('%Women in Education field')

# Sử dụng plt.tight_layout() trước khi show()
plt.tight_layout() #Để căn chỉnh các đồ thị sao cho label hay title của chúng không bị overlap nhau
plt.show()

# Bạn nhận xét gì về biểu đồ vừa vẽ ====> Physics tăng đều, CS tăng rồi giảm, 
#                                         Health thì biến động lên xuống mạnh, nhưng nhìn chung vẫn tăng. 
#                                         Education tăng giống Health, nhưng biến động lên xuống thường xuyên hơn
```
