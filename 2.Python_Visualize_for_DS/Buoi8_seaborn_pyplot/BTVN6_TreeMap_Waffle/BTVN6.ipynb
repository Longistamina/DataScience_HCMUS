```python
# Vẽ TreeMap thể hiện tỷ lệ số phiếu bầu lần lượt cho ứng viên ở Virginia
plt.figure(figsize=(8,6))
squarify.plot(sizes=df.Virginia, label=df.Name, #sizes là đưa số liệu cần visualized vào, label là tên ứng với số liệu đó
             color=['#a9ac6e','#9f5956','#74e5eb']) 
                                                #Dựa trên tương quan về  số liệu mà các ô sẽ có diện tích tượng trưng, thể hiện ít nhiều
plt.axis('off') #Để tắt đi trục kẻ vạch, nhìn cho đơn giản
plt.title('2016 Virginia Presidential Election Results')
plt.show()
```

### Part 2: Waffle Chart

```python
# Tính tổng số phiếu bầu của từng ứng viên ở cả 3 khu vực
df.drop(columns='Name').sum(axis=1) #Loại bỏ cột Name, vì cột name chỉ chứa string, những gì còn lại là int thì tính tổng
                                    #axis = 1 để  tính theo hàng
```

```python
#Thêm cột total chỉ tổng số  phiếu vào df ban đầu
df['Total'] = df.drop(columns='Name').sum(axis=1).tolist() #Phải đổi tập dữ liệu về dạng list rồi mới nhét vào cột mới được
df
```
