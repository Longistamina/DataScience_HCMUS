## Visualizing stock trends.

Giả sử bạn muốn đầu tư vào chứng khoán và bạn đã download giá đóng cửa - mở cửa của 5 "ông lớn" sau: Amazon, Google, Apple, Facebook, and Microsoft. 
Hãy tạo đồ thị line để quan sát xu hướng chứng khoán của các công ty trên. 


```python
# Import thư viện
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Sử dụng thư viện pandas để đọc dữ liệu

```python
# load dữ liệu
apple = pd.read_csv('data/AAPL_data.csv')
amazon = pd.read_csv('data/AMZN_data.csv')
fb = pd.read_csv('data/FB_data.csv')
ggle = pd.read_csv('data/GOOGL_data.csv')
mcrsoft = pd.read_csv('data/MSFT_data.csv')
```

Sử dụng Matplotlib để tạo một biểu đồ trực quan hóa giá đóng cửa trong 5 năm qua (toàn bộ dữ liệu trong file) cho cả 5 công ty. Lưu ý có nhãn, tiêu đề và legend để làm cho đồ thị dễ xem hơn và sử dụng plt.grid () cho đồ thị.

```python
# Tạo figure (figure size of 16"x8")
plt.figure(figsize=(16,8))

# Vẽ đồ thị
plt.plot(apple.date, apple.close, color='grey', label='Apple')
plt.plot(amazon.date, amazon.close, color='yellow', label='Amazon')
plt.plot(fb.date, fb.close, color='blue', label='Facebook')
plt.plot(ggle.date, ggle.close, color='green', label='Google')
plt.plot(mcrsoft.date, mcrsoft.close, color='orange', label='Microsoft')

# Đặt ticks cho trục x- và trục y
plt.xticks(np.arange(0, 1260, 40), rotation=70)
plt.yticks(np.arange(0, 1450, 100))

# Title and labels
plt.xlabel('MỐC THỜI GIAN',size=15)
plt.ylabel('GIÁ ĐÓNG CỬA (USD)',size=15)
plt.title('GIÁ ĐÓNG CỬA CỦA CỔ PHIẾU "5 ÔNG LỚN" CÔNG NGHỆ MỸ THEO THỜI GIAN', color='brown')

#  grid
plt.grid(alpha=0.5)

#  legend
plt.legend()

# Show plot
plt.show()
```
