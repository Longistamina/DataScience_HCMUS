```python
drks.groupby('continent')[['beer_servings','spirit_servings','wine_servings','total_litres_of_pure_alcohol']].median()
#Tính trung vị tất cả số lượng rượu bia tiêu thụ theo châu lục
```

```python
drks.groupby('continent').beer_servings.agg(['mean','median','max']) #Gộp nhiều hàm tính mean, median và max cho beer_servings theo châu lục
```

```python
drks.sort_values('beer_servings') #Sắp xếp tăng dần theo beer_servings
```

```python
drks[drks['beer_servings']>0].sort_values('beer_servings') #Sắp xếp tăng dần theo beer_servings, chỉ lấy những giá trị >0
```

```python
drks[drks['beer_servings']>0].sort_values(by='beer_servings',ascending=False) #Sắp xếp giảm dần theo beer_servings, chỉ lấy những giá trị >0
```

```python
drks[drks['beer_servings']>0].sort_values('beer_servings').head() #5 quốc gia tiêu thụ bia lớn nhất
```

```python
drks[drks['beer_servings']>0].sort_values(by='beer_servings',ascending=False).tail() #5 quốc gia tiêu thụ bia ít nhất
#Xếp giảm dần rồi lấy tail()
```
