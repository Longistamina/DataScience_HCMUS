```python
#----------------------------THÊM CỘT results, nếu open>close thì điền 'up', còn open<close thì điền 'down'---------------------#
```

```python
stk_cpns['result'] = np.where(stk_cpns['open']>stk_cpns['close'],'up','down')
#Hàm np.where sẽ so sánh giá trị open có lớn hơn close hay không ở từng hàng, nếu TRUE thì điền 'up' vào result, FALSE thì điền down
```

```python
stk_cpns.head()
```

```python
#Cách làm khác, dùng hàm .loc để  truy cập vào ô toạ độ (hàng,cột) cụ thể
#Đi tham chiếu từng dòng cụ thể. Cột thị sẽ cố định ở 'result', dòng nào thoả điều kiện nào thì điền giá trị ứng với điều kiện đó vào result

stk_cpns.loc[stk_cpns['open']>stk_cpns['close'],'result'] = 'up'
stk_cpns.loc[stk_cpns['open']<=stk_cpns['close'],'result'] = 'down'
```

```python
stk_cpns.head()
```

```python
#Cách khác nữa là dùng apply hoặc map và lambda
stk_cpns['result'] = (stk_cpns['close'] < stk_cpns['open']).apply(lambda x: 'down' if x else 'up')
stk_cpns['result'] = (stk_cpns['close'] < stk_cpns['open']).map(lambda x: 'down' if x else 'up')

# x là kết quả của phép so sánh stk_cpns['close'] < stk_cpns['open']
# nếu x là TRUE, tức close<open thì điền down
# ngược lại (else) thì điền up
#LƯU Ý: map() chỉ dùng được cho series(1 chiều) chứ không dùng cho data frame (2 chiều) được, apply() dùng được cho cả hai
```

```python
#Cách khác cũng dùng apply
stk_cpns['result'] = stk_cpns.apply(lambda x: 'up' if x['close'] > x['open'] else 'down', axis = 1)

#Khi viết stk_cpns.apply(lambda x:...) thì nó sẽ hiểu x là dò từng cột của df stk_cpns để  thực hiện yêu cầu
# nhưng phía sau có ghi axis=1 thì lúc này nó sẽ hiểu x là dò từng dòng của df stk_cpns để  thực hiện yêu cầu
```
