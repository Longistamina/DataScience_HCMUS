```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
f = open('/home/long/Documents/Data Science/2. Python for DS/file of Bảng tin/heights_1.txt')
data = f.read()
f.close

lst_hei = eval(data)
lst_hei
```

```python
df_h=pd.DataFrame({'Heights_inch':lst_hei})
df_h.head()
```

```python
df_h['Heights_m']=df_h['Heights_inch']*0.0254 #Thêm cột Heights_m vào, giá trị h_m bằng h_inch*0.0254
df_h.head()
```

```python
df_h['Heights_m'].plot.box(notch=True) #Chỉ lấy cột Heights_m để vẽ, không đụng đến các cột khác
```
