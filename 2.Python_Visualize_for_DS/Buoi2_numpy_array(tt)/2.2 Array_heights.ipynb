```python
import numpy as np
```

```python
#Đọc tập tin heights_1.txt vào dưới dạng array (chiều cao của vđv bóng rổ, đơn vị là inch)
f = open('heights_1.txt')
data_heights = f.read() #Đọc dữ liệu trong tập tin, trả về dạng string
f.close
print(data_heights)
```

```python
lst_heights = eval(data_heights) #eval() đánh giá chuỗi heights, thấy giống list nên convert sang list luôn
```

```python
arr_heights = np.array(lst_heights) #CHuyển lst thành array
```

```python
arr_heights.shape
```

```python
arr_heights[0:10] #Xem 10 người đầu tiên, từ 0 đến cận 10 (là 9)
```

```python
arr_heights[-10:] #Xem 10 người cuối cùng
```

```python
#Thống kê đơn giản
print('Average heights:',np.mean(arr_heights))  #Không theo hướng đối tượng, dùng function
print('Average heights:',arr_heights.mean())   #Theo kiểu hướng đối tượng (Object oriented), dùng method
```

```python
#Đổi chiều cao từ inch sang mét (1 inch = 0.0254 mét)
arr_met = arr_heights*0.0254
arr_met
```

```python
print('mean:',arr_met.mean())
print('max:',arr_met.max())
print('min:',arr_met.min())
print('median:',np.median(arr_met)) #Vì class array không có method median nên phải dùng hàm numpy.median()
```

```python
arr_met[arr_met <= 1.8] #Truy xuất chiều cao <= 1.8 mét
```

```python
arr_met[arr_met > 2.0] #Truy xuất chiều cao > 2 mét
```

```python
#Toán tử  and: &
#Toán tử  or: |
#Toán tử not: ~

arr_met[(arr_met < 2.0) & (arr_met > 1.8)]
```

```python
arr_met5 = arr_met[0:5] #tạo mảng gồm 5 giá trị đầu của arr_met
arr_met5
```

```python
print(arr_met5)
print(np.append(arr_met5,1.93)) #Thêm giá trị 1.93 vào cuối arr_met5
```

```python
print(arr_met5)
print(np.append(arr_met5,[1.75,1.82,1.91])) #Thêm 3 giá trị trên vào cuối arr_met5
```

```python
print(arr_met5)
print(np.insert(arr_met5,1,1.65)) #Thêm giá trị 1.65 vào arr_met5 ở index = 1
```

```python
print(arr_met5)
print(np.delete(arr_met5,2)) #Xoá giá trị có index = 2 của arr_met5
```
