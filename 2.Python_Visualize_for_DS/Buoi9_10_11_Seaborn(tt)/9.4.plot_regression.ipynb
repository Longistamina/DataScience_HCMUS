```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.regplot(data=tips, x='total_bill', y='tip'); # y = ax + b
                                                 #Cả x và y đều là hai biến định lượng
```

```python
sns.regplot(data=tips, x='total_bill', y='tip', ci=None); #ci=None để  không hiển thị miền khoảng tin cậy mờ mờ
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex'); # lm = linear model, vẽ đường hồi quy tuyến tính
                                                           #hue='sex' để  phân nhóm dữ liệu theo giới tính, 2 đường có màu khác nhau
                                                           #Chỉ có lmplot mới có argument hue
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue='smoker');
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue='smoker', col='time'); #col để  chia thành nhiều biểu đồ con
                                                                          #mỗi biểu đồ nằm trên một cột, theo time
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue='smoker', row='time'); #row để  chia thành nhiều biểu đồ con
                                                                          #mỗi biểu đồ nằm trên một hàng, theo time
```
