```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
sns.jointplot(data=iris, x='petal_length', y='petal_width', kind='reg'); # kind: "scatter" | "reg" | "resid" | "kde" | "hex"
```

```python
sns.catplot(data=iris, x='species', y='petal_length');
# kind: "point", "bar", "strip", "swarm", "box", "violin", or "boxen"
```

```python
sns.catplot(data=iris, x='species', y='petal_length', kind='bar');
```
