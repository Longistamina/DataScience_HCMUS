```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# du lieu titanic, ve bieu do tan suat cua fare
titanic = sns.load_dataset('titanic')
```

```python
titanic.shape
```

```python
titanic.head()
```

```python
sns.displot(titanic.fare, kind='hist'); #Distplot để vẽ đồ thị tần suất phân phối distribution của tập dữ liệu
```

```python
sns.displot(titanic.fare, rug=True, color='magenta',kind='kde'); #kind='kde' để vẽ đường kernel density estimate
                                        #rug = True để  thay kiểu hiển thị thành dạng trải thảm, chỗ  nào càng đậm là dữ liệu tại đó nhiều
```
