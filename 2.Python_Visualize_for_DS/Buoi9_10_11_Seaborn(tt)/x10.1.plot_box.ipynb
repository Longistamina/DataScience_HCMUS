```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.boxplot(data=titanic, x='survived', y='age'); #Vẽ boxplot với biến định lượng là Age, biến phân loại là Survived
```

```python
sns.boxplot(data=titanic, x='survived', y='age');
sns.swarmplot(data=titanic, x='survived', y='age', palette='rainbow'); #Vẽ swamplot chồng lên boxplot
                                                                       #palette='rainbow' để  tạo màu khác cho swamplot, không trùng
```

```python
sns.boxplot(data=titanic, x='pclass', y='age'); #Vẽ boxplot với biến định lượng là Age, biến phân loại là Pclass
                                                #Để thấy độ tuổi của từng khoang hạng
```

```python
sns.boxplot(data=titanic, x='age', y='pclass', orient='h'); #dùng orient='h' để  vẽ kiểu ngang horizontal
                                                            #Tên trục cũng phải đảo ngược
```
