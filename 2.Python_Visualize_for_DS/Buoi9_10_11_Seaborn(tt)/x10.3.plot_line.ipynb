```python
museum_filepath = "vd_seaborn/museum_visitors.csv"
museum_data = pd.read_csv(museum_filepath, index_col="Date", parse_dates=True)
#Chọn cột Date làm cột index
#Vì trong dữ liệu có ngày tháng năm, nên bật parse_dates=True để nó hiểu đó là datetime, chứ không phải string thông thường
```

```python
museum_data.shape
```

```python
museum_data.info()
```

```python
museum_data.head()
```

```python
# Set the width and height of the figure
plt.figure(figsize=(14,6))

# Line chart showing the number of visitors to each museum over time
sns.lineplot(data=museum_data); #Vẽ biểu đồ lineplot thể hiện lượng khách theo thời gian của các viện bảo tàng

#Nhận xét: viện bảo tàng đường màu xanh lá cây và màu đỏ lượng khác thăm không nhiều, nhưng ổn định theo tgian
#          viện bảo tàng màu xanh dương thì luôn cao và ổn định
#          viện bảo tàng đường màu cam lúc đầu tăng vọt, sau đó giảm dần rồi ổn định
#          Một số  viện có chu kỳ lên xuống, có thể theo mùa....
```

```python
#Chỉ vẽ lineplot cho một viện bảo tàng
sns.lineplot(data=museum_data['Avila Adobe'], label='Avila Adobe')
plt.xlabel('Date');
```
