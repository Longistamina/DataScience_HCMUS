```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
emp = pd.read_excel('/home/long/Documents/Data Science/2. Python for DS/file of Bảng tin/employees.xls')
```

```python
emp.head()
```

```python
str_date = emp.HIRE_DATE.str.split('-') #Chuyển hết Hire_date về dạng string, rồi split chúng ra bằng '-'
working_years = [] #Tạo list chứa thâm niên làm việc, tính bằng năm 2024 trừ đi năm được thuê (hire_date)

for i in range(len(str_date)): #Duyệt qua từng list con trong list str_date
    working_years.append(2024 - int(str_date[i][0])) #Lấy phần tử đầu trong list con thứ 'i' là năm, chuyển sang int
                                                     # Rồi lấy 2024 trừ cho số năm đó để  ra được thâm niên làm việc, cho vào working_years
print(working_years)
```

```python
sbn.scatterplot(x=emp.SALARY,y=working_years, #Vẽ scatterplot để thấy tương quan giữa salary và working_years
                hue=emp.JOB_ID); #Phân loại dữ liệu theo JOB-ID, tức mỗi JOB_ID khác nhau sẽ được tô màu khác nhau
```

```python
sbn.relplot(x=emp.SALARY, y=working_years, hue=emp.JOB_ID, col=emp.DEPARTMENT_ID);
#Dùng relplot (relation plot) để vẽ thì có thể chia thành nhiều đồ thị nhỏ khác nhau
#col=emp.DEPARTMENT_ID tức mỗi nhóm dữ liệu thuộc cùng một DEPARTMENT_ID sẽ được vẽ trên cùng một biểu đồ
#col nên mỗi biểu đồ nằm trên một cột khác nhau
```

```python
sbn.relplot(x=emp.SALARY, y=working_years, hue=emp.JOB_ID, row=emp.DEPARTMENT_ID);
#giống như trên, nhưng vì dùng row=emp.DEPARTMENT_ID nên mỗi biểu đồ  của 1 depart_ID sẽ nằm trên một hàng khác nhau
```
