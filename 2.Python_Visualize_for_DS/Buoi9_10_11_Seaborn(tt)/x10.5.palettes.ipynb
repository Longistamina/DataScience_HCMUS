```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.palettes.SEABORN_PALETTES # red, green, blue (một chuỗi 6 ký tự, 2 ký tự đầu - Red, 2 ký tự giữa - Green, 2 ký tự cuối - Blue)
                              # Cho biết danh sách palette hiện có của palette
```

```python
iris = sns.load_dataset('iris')
```

```python
for p in sns.palettes.SEABORN_PALETTES: #Vẽ đồ thị với các bộ palette khác nhau bằng vòng lặp for
    sns.set_palette(p)
    sns.distplot(iris['petal_length'])
    plt.show()
```

```python
for p in sns.palettes.SEABORN_PALETTES:
    sns.set_palette(p)
    sns.swarmplot(x="species", y="petal_length", data=iris);
    plt.show()
```

```python
current_palette = sns.color_palette()
sns.palplot(current_palette)
plt.show()
```

```python
for p in sns.palettes.SEABORN_PALETTES: #Dùng câu lệnh này để  vẽ ra hết màu của các pallette đi kèm tên
    sns.set_palette(p)
    print(p) #in ra tên của palette màu hiện hành
    sns.palplot(sns.color_palette()) #Vẽ đồ thị bằng palette hiện hành
    plt.show()
```

```python
# Sequential Colors - Màu tuần tự, dùng cho những dữ liệu có ý nghĩa sắp xếp, tăng dần giảm dần
sns.palplot(sns.color_palette('Reds', 12))
```

```python
p = sns.light_palette("brown", reverse=False) #Reverse=False thì palette màu chạy từ nhạt sang đậm
sns.palplot(p)
```

```python
p = sns.light_palette("brown", reverse=True) #Reverse=True thì palette màu chạy ngược lại từ đậm về nhạt
sns.palplot(p)
```

```python
df = data.groupby('DEPARTMENT_ID', as_index=False).EMPLOYEE_ID.count().sort_values(by='EMPLOYEE_ID')
df #Tạo df cho biết ứng với từng Department_ID thì có bao nhiêu nhân viên, xếp tăng dần
```

```python
sns.set_palette(sns.color_palette('Reds', 12)) #Vẽ màu tuần tự, phòng ID nào càng có nhiều nhân viên thì màu càng đậm
sns.barplot(x='DEPARTMENT_ID', y='EMPLOYEE_ID', data=df, ci=None, order=df['DEPARTMENT_ID'])
plt.show()
```

```python
df = df.sort_values(by='EMPLOYEE_ID', ascending=False)
df #Xếp giảm dần df trên
```

```python
sns.set_palette(sns.color_palette('Reds', 12))
sns.barplot(x='DEPARTMENT_ID', y='EMPLOYEE_ID', data=df, ci=None, order=df['DEPARTMENT_ID'])
plt.show()
```

```python
sns.set_palette(reversed(sns.color_palette('Reds', 12)), 12) #Có reversed để đảo ngược palette màu
sns.barplot(x='DEPARTMENT_ID', y='EMPLOYEE_ID', data=df, ci=None, order=df['DEPARTMENT_ID'])
plt.show()
```
