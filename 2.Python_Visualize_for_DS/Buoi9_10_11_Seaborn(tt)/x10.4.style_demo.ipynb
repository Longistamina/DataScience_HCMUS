```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
sns.set() #Câu lệnh này dùng để  chỉ định style vẽ biểu đồ  theo seaborn
          #Nghĩa là không cần dùng lệnh trong thư viện seaborn nhưng vẫn vẽ được đồ thị theo kiểu của seaborn
```

```python
iris['petal_length'].plot.hist(); #Lệnh .plot.hist() này không thuộc thư viện seaborn,
                                  #Nhưng do trước đó áp đặt style theo seaborn nên được vẽ theo kiểu seaborn
```

```python
sns.set_style('whitegrid') #Vẽ theo style là whitegrid
```

```python
iris['petal_length'].plot.hist();
```

```python
sns.distplot(iris['petal_length']);
```

```python
for style in ['white','dark','whitegrid','darkgrid','ticks']:
    sns.set_style(style)
    sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species')
    plt.show()
```

```python
sns.set_palette('Dark2') #Áp đặt màu sẵn lên đồ thị
ax = sns.swarmplot(x='species', y='petal_length', data=iris) # axes
ax.set_title('Iris types', fontsize=14, color='blue')
plt.show()
```

```python
p = sns.color_palette('pastel') #Gán màu cần dùng vào một biến, rồi gọi ra biến đó dùng sau
ax = sns.boxplot(x='species', y='petal_length', data=iris, palette=p) #Phải set tham số  palette
ax.set_title('Iris types', fontsize=14, color='red')
plt.show()
```
