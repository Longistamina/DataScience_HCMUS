```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.stripplot(data=tips, x='time', y='total_bill',hue='time'); #Vẽ stripplot với biến category là time, và biến quantitative là total_bill
```

```python
sns.stripplot(data=tips, x='time', y='total_bill', hue='sex'); #hue=sex để  tô màu dữ liệu khác nhau theo giối tính
```

```python
sns.barplot(data=tips, x='time', y='total_bill', hue='sex', estimator=sum); #Vẽ barplot bằng giá trị sum của total_bill, dùng estimator=sum
                                                                            #mặc định nó sẽ để estimator=mean, tức dùng mean để vẽ barplot
```

```python
sns.stripplot(data=tips, y='time', x='total_bill', hue='sex'); #Bản đồ stripplot rất dễ xoay, chỉ cần đổi y="time", x="total_bill"
```

```python
sns.barplot(data=tips, y='time', x='total_bill', hue='sex', estimator=sum); #Barplot cũng có thể xoay tương tự như thế, với seaborn
```

```python
sns.swarmplot(data=tips, x='time', y='total_bill', hue='sex'); #Swarmplot là dạng biểu đồ  tương tự stripplot
                                                               # Chỉ khác là các dữ liệu không chồng chéo lên nhau
                                                               #Dùng swarmplot thay cho stripplot khi dữ liệu quá nhiều
```

```python
titanic.info() #Xem info để biết biến nào là category, biến nào là quantitative để chọn vẽ cho đúng
```

```python
titanic.pclass.unique() #Để  xem xem trong cột pclass có những loại giá trị gì
```

```python
# ve cac quan sat voi x=class, y=age
sns.swarmplot(data=titanic, x='class', y='age', hue='sex');
```

```python
sns.countplot(data=titanic, x='class', hue='sex'); #Đếm xem trong mỗi class có bao nhiêu quan sát, trong đó bao nhiêu male / female
```

```python
# ve grid
sns.catplot(data=titanic, x='class', y='age', col='sex', kind='swarm', hue='class');
```

```python
#Vẽ boxplot và swarm plot chồng lên nhau (cái nào xuất hiện trước thì nằm ở trên)
#Dùng pastel cho boxplot để dễ nhìn, không trùng với swarmplot

sns.swarmplot(data=titanic, x='class', y='age', hue='class');
sns.boxplot(data=titanic, x='class', y='age', palette='pastel',hue='class');
```
