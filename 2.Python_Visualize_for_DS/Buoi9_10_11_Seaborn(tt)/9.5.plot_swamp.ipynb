```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore') #Câu lệnh này để ignore những warning mà pandas đưa ra cho câu lệnh

#Swarm plot cũng là dạng scatter, nhưng vẽ theo từng cụm dữ liệu khác nhau, trông giống tổ ong
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.shape
```

```python
iris.info()
```

```python
iris.species.unique().tolist() #Kiểm tra xem cột species của dataset iris có bao nhiêu loại giá trị, xuất kq ra thành list
```

```python
sns.swarmplot(data=iris, x='species', y='petal_length'); #Vẽ swarm plot thể hiện petal length của cả 3 loài hoa
                                                         #x là biến phân loại species
                                                         #y là biến định lượng petal_length
```

```python
sns.scatterplot(data=iris, x='petal_length', y='petal_width'); 
#Vẽ scatterplot để thể hiện tương quan giữa petal width và length của toàn bộ các cây thuộc chi iris
```

```python
sns.regplot(data=iris, x='petal_length', y='petal_width'); #Vẽ thêm đường hồi quy tuyến tính giữa petal width và length
```

```python
# muon biet cac hanh khach tren tau titanic co cac tuoi theo pclass
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.swarmplot(data=titanic, x='pclass', y='age'); #Vẽ swarm plot với x(category) là pclass, còn y(quantitative) là tuổi
```

```python
sns.swarmplot(data=titanic, x='pclass', y='age', hue='sex'); #Thêm hue='sex' để thấy rõ phân bố dữ liệu theo giới tính
```

```python
sns.stripplot(data=titanic, x='pclass', y='age'); #Trong trường hợp dữ liệu phân nhóm nhiều quá thì nên dùng striplot thay cho swamplot
```
