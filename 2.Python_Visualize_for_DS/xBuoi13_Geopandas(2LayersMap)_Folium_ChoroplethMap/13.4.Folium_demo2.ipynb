```python
import folium #pip3 install folium
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#-----------------Marker dùng để đánh dấu hay ký hiệu các vị trí trên bản đồ lấy từ Folium-------------#
```

```python
#Đọc dữ liệu trong file HCM_location.xlsx, đọc sheet "Location" để lấy thông tin các quận
df_hcm = pd.read_excel('data/HCMC_location.xlsx', sheet_name='Location')
df_hcm.head()
```

```python
#Tạo bản đồ thành phố HCM
hcm_latitude = 10.762622
hcm_longitude = 106.660172
hcm_map = folium.Map(location=[hcm_latitude, hcm_longitude], zoom_start=12)
hcm_map
```

```python
#Thêm các Circle Marker để  cho thấy vị trí các quận trên Tp. HCM

#Instantiate a feature group for the districts in the df_hcm, so that we can put things in it and handle them as a single layer
districts = folium.map.FeatureGroup()

#Loop through the districts and add each folium.CircleMarker to the district feature group
for lat,lng in zip(df_hcm.Latitude, df_hcm.Longitude):
    districts.add_child(
        folium.CircleMarker(
            [lat,lng],
            radius=5, #Bán kính của circle marker
            color='yellow',
            fill=True,
            fill_color='red',
            fill_opacity=0.6
        )
    )

#add the districst to hcm map
hcm_map.add_child(districts)

#Tư duy của cách làm này là vẽ các CircleMarker trên một layer riêng (districs~FeatureGroup), với toạ độ trùng với toạ độ các quận
#Sau đó chồng layer này lên layer hcm_map là ra được bản đồ tp.HCM có các CircleMarker đánh dấu các quận
```

```python
#-------------Tạo thêm Pop-up marker cho các quận của HCM-------------#

#hcm_map2 = folium.Map(location=[hcm_latitude, hcm_longitude], zoom_start=12)

for lat,lng,name in zip(df_hcm.Latitude, df_hcm.Longitude,df_hcm.Name):
    folium.Marker([lat,lng], popup=name).add_to(hcm_map) #Có thể add thẳng vào hcm_map như thế này mà không cần thông qua FeatureGroup
    #popup=name để khi click vào popup trên bản đồ thì tên quận hiện ra

hcm_map
```

```python
#-----------___Thêm thông tin dân số  (population) vào trong map---------------#

hcm_map = folium.Map(location=[hcm_latitude, hcm_longitude], zoom_start=12)
# tạo các Popup
# add pop-up text to each marker on the map
latitudes = list(df_hcm.Latitude)
longitudes = list(df_hcm.Longitude)
labels = list(df_hcm.Name)
pops = list(df_hcm.Population)

for lat, lng, label, pop in zip(latitudes, longitudes, labels, pops):
    tt = 'So dan: '+str(pop)
    folium.Marker([lat, lng], popup=label, tooltip=tt).add_to(hcm_map)    
hcm_map
```
