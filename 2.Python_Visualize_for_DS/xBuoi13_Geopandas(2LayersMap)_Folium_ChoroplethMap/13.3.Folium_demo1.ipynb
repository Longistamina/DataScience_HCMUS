```python
import folium #pip3 install folium
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#-----------------Folium là một cơ sở dữ liệu cho phép truy cập và lấy địa điểm nào đó trên bản đồ thế giới để vẽ Leaflet map-------------#
```

```python
#Hiển thị bản đồ thế giới
world_map = folium.Map()
world_map
```

```python
#Tạo bản đồ  với  center là nước Canada (Location=[56.130, -106.35] và zoom level là zoom_start=4
canada_map = folium.Map(location=[56.130, -106.35], zoom_start=4)
canada_map
```

```python
#Tạo bản đồ thành phố HCM
hcm_latitude = 10.762622
hcm_longitude = 106.660172
hcm_map = folium.Map(location=[hcm_latitude, hcm_longitude], zoom_start=12)
hcm_map
```

```python
#Tạo bản đồ thành phố HCM với style vẽ là "Stadia.StamenTerrain"
hcm_latitude = 10.762622
hcm_longitude = 106.660172
hcm_map2 = folium.Map(location=[hcm_latitude, hcm_longitude], zoom_start=12, tiles="Stadia.StamenTerrain")
hcm_map2
```
