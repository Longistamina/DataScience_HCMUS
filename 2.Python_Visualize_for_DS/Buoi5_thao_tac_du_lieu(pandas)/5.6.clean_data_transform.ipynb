```python
stocks.loc[3,'high'] = h_mean #Thay giá trị NaN thứ nhất ở hàng index 3 và cột 'high' bằng giá trị h_mean
stocks.loc[12,'high'] = h_mean
```

```python
l_mean = stocks['low'].mean()
l_mean
```

```python
stocks.loc[3,'low'] = l_mean
stocks.loc[12,'low'] = l_mean
```

```python
#---------------------Phương án 2: dùng symbol để  tham chiếu rồi thay NaN bằng mean-------------------------#
```

```python
stocks2 = pd.read_csv('/home/long/Documents/Data Science/2. Python for DS/file of Bảng tin/stocks1(22).csv')
```

```python
stocks2.groupby('symbol').high.mean() #Nhóm các dữ liệu cột high theo symbol, rồi tính mean của từng nhóm
```

```python
stocks2.groupby(['symbol'])['high'].transform('mean')
#Câu lệnh này đầu tiên sẽ nhóm dữ liệu trong cột high theo symbol
#sau đó, nó sẽ tính mean từng nhóm symbol
#rồi nó sẽ dò từng dòng xem dòng này có symbol là gì, rồi lấp giá trị mean ứng với symbol đó vào cột high
#Ví dụ, ở hàng có index = 0, có symbol là AMZN, thì mean của cột high theo nhóm AMZN là 1690.310,
#  nên hàm transform sẽ chuyển giá trị cột high của hàng 0 này là 1674.26 thành giá trị mean tương ứng với nhóm AMZN là 1690.31
```

```python
stocks2['high'].fillna(stocks2.groupby(['symbol'])['high'].transform('mean'), inplace = True)
#Cũng tính toán như trên nhưng lúc này chỉ transform thành giá trị high mean nhóm theo symbol ở những chỗ  nào có NaN
#Đồng thời cập nhật dữ liệu lên thẳng df stocks2 luôn
```

```python
stocks2['low'].fillna(stocks2.groupby(['symbol'])['low'].transform('mean'), inplace = True) #Làm tương tự với low
```
