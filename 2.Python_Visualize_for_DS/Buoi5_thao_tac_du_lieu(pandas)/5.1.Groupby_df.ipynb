```python
import numpy as np
import pandas as pd
```

```python
drinks  = pd.read_csv("/home/long/Documents/Data Science/2. Python for DS/file of Bảng tin/drinks.csv", index_col=0)
```

```python
drinks.head()
```

```python
drinks.groupby('continent').beer_servings.mean() #Xếp nhóm dữ liệu theo continent, rồi tính mean của beer_servings của từng continent
```

```python
drinks.drop(columns='country').groupby("continent").mean() #Loại bỏ cột country là non-numeric, 
                                                           #Tính mean tất cả các cột numeric còn lại, phân loại theo continent
```

```python
drinks.drop(columns='country').groupby("continent").agg(['sum','mean','min','max']) #Bỏ cột non-numeric là 'country'
                                                                                    #Thực hiện nhiều thống kê khác nhau
                                                                                    #Phân loại theo continent
```

```python
drinks.drop(columns='country').groupby("continent").beer_servings.agg(['sum','mean','min','max'])
```
