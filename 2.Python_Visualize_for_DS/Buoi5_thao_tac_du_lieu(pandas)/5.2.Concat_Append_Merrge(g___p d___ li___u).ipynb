```python
emp1 = emp[['EMPLOYEE_ID', 'FIRST_NAME', 'SALARY', 'DEPARTMENT_ID']] #Chỉ lấy các cột có tên tương ứng ra, tạo thành một df mới là emp1
```

```python
df_emp1_dpmt = pd.merge(left=emp1, right=dpmt, left_on='DEPARTMENT_ID', right_on='DEPARTMENT_ID')
#Merge emp1 và dpmt lại với nhau
#Trong đó, emp1 nằm bên trái, còn dpmt nằm bên phải
#Cột DEPARTMEN_ID của hai dataframe sẽ được so sánh đối chiếu nhau sao cho khớp, và nhập thành 1
```

```python
df_emp1_dpmt = pd.merge(emp1,dpmt)
#Cách này ghi là cách ghi gọn hơn
#Dataframe nào được truyền vào trước thì sẽ nằm bên trái (left), cái còn lại nằm bên phải (right)
#Nếu left_on và right_on không được chỉ định thì nó sẽ là None. Lúc này, pd.merge sẽ tự xem cột nào giữa hai df trùng nhau để  khớp dữ liệu
#Tuy nhiên, trong trường hợp hai df có từ hai cột trở lên trùng nhau thì nên chỉ rõ left_on là gì, right_on là gì để  merge cho chính xác
```

```python
#So khớp bằng (how = 'inner'): là trong cột dùng để  khớp (vd: cột DEPARTMENT_ID), cả hai bên đều phải có giá trị tương ứng cho nhau
#ví dụ bên df1 có các giá trị 1 2 3 4 5 thì cột còn lại cũng phải có các giá trị 1 2 3 4 5 để so khớp, không hơn không kém
#hàng nào của một trong hai bên không chứa giá trị nào mà chỉ có null thì sau khi gộp xong, hàng đó sẽ bị mất đi

#So khớp ưu tiên: trong trường hợp cột dùng để  so khớp giữa hai bảng có những giá trị tương ứng nhau, thì phải chọn ra bên ưu tiên để gộp
#ví dụ bên df1 có các giá trị 1 2 3 4 5 nhưng bên df2 cũng cột đấy nhưng chỉ có 1 2 3 4 thì có thể chọn một bên để  ưu tiên so khớp
#ví dụ chọn ưu tiên xếp theo df2, khi đó giả sử  hàng nào của df2 không có 1 2 3 4 5, chỉ có null thì sau khi gộp vẫn đượcc giữ lại 
```
