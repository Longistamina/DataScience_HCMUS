```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#-------------------------LINE PLOT: biểu đồ đường-----------------------------#
```

```python
year = [1950,1951,1952,1953,1954]
pop = [2.53, 2.57, 2.62, 2.67, 2.71] #pop = population = dân số
```

```python
plt.plot(year,pop)  #Tạo object là một biểu đồ đường
                    #Truyền dữ liệu trục x vào trước (là year)
                   #Sau đó truyền dữ liệu trục y vào sau (pop)

plt.show() #Hiện line plot mới vẽ
```

```python
plt.plot(year,pop)     #Tạo object là một biểu đồ đường, trục X là year, trục Y là population
plt.xticks(year,year) #Điều chỉnh lại dữ liệu year trên biểu đồ thành dữ liệu year được truyền vào, thành integer không có chấm thập phân
plt.show()
```

```python
plt.plot(year,pop)    ##Tạo object là một biểu đồ đường, trục X là year, trục Y là population
plt.xticks(year,year) #Điều chỉnh lại dữ liệu year trên biểu đồ thành dữ liệu year được truyền vào, thành integer không có chấm thập phân
plt.xlabel('Năm')     #Thêm tên trục X
plt.ylabel('Dân số  (tỷ người)') #Thêm tên trục Y
plt.title('DÂN SỐ  THẾ  GIỚI QUA CÁC NĂM', color='r', fontsize=14) #Thêm tên biểu đồ 
plt.show()
```

```python
plt.plot(year,pop, 'bo')  #Chữ  'bo' để  nó hiểu là vẽ chấm thôi chứ không vẽ đường
plt.xticks(year,year) #Điều chỉnh lại dữ liệu year trên biểu đồ thành dữ liệu year được truyền vào, thành integer không có chấm thập phân
plt.xlabel('Năm')     #Thêm tên trục X
plt.ylabel('Dân số  (tỷ người)') #Thêm tên trục Y
plt.title('DÂN SỐ  THẾ  GIỚI QUA CÁC NĂM', color='r', fontsize=14) #Thêm tên biểu đồ 
plt.show()
```

```python
plt.figure(figsize=(8,5)) #Figure là khung hình bao xung quanh toàn bộ biểu đồ
                          # figsize=(8,5) tức chiêu ngang 8 inch, chiều cao 5 inch

plt.plot(year,pop, color='magenta', marker='o', linestyle='dashed',   #marker để định dạng chấm, markersize là kích thước chấm
        linewidth=2, markersize=12)                                   #linestyle để định dạng đường nối giữa các chấm
                                                                     
plt.xticks(year,year) 
plt.grid(True) #Câu lệnh này để  hiện ra lưới ô vuông
plt.xlabel('Năm')     
plt.ylabel('Dân số  (tỷ người)') 
plt.title('DÂN SỐ  THẾ  GIỚI QUA CÁC NĂM', color='r', fontsize=14) 
plt.show()
```

```python
#Vẽ biểu đồ đường biểu thì điểm trung bình môn toán theo các lớp 10A1, 10A2, 10A3, 10A4 và 10A5
df_lop_dtbT = pd.DataFrame({'lop':['10A1', '10A2', '10A3', '10A4', '10A5'],
                           'dtbToan':[8.7, 6.5, 9.2, 7.3, 6.7]})
df_lop_dtbT
```

```python
plt.figure(figsize=(8,5))
plt.plot(df_lop_dtbT['lop'], df_lop_dtbT['dtbToan'], marker='o', linestyle='dashed', color='green')
plt.xlabel('Lớp', size=15)
plt.ylabel('Điểm trung bình môn Toán', size=15)
plt.ylim(0,10)
#Hoặc plt.yticks(np.arrange(11))
plt.title('Biểu đồ điểm trung bình môn Toán của khối lớp 10', color='r')
plt.show
```

```python
#Vẽ hai đường trên một biểu đồ
#Hai học có các điểm Toán Lý Hoá khác nhau.

df_diem_hs = pd.DataFrame({'Toán':[8.5,7.5],
                          'Lý':[7.5,6.5],
                          'Hoá':[7.8,8.7],},
                         index=['An','Bình'])
df_diem_hs
```

```python
plt.figure(figsize=(8,5))

plt.plot(df_diem_hs.columns, df_diem_hs.loc['An'], marker='o', linestyle='dashed', color='blue', label='An') 
plt.plot(df_diem_hs.columns, df_diem_hs.loc['Bình'], marker='X', linestyle='dashed', color='green', label='Bình')
#Truyền vào trục X tên tất cả các cột của dataframe, bằng câu lệnh df_diem_hs.columns
#Truyền vào trục Y giá trị điểm ứng với index hàng là tên của học sinh, bằng câu lệnh df_diem_hs.loc['An']

plt.legend() #Thêm bảng chú thích
plt.show()
```

```python
#Thêm miền area bên dưới biểu đồ đường:
```

```python
plt.figure(figsize=(8,5))
plt.plot(df_lop_dtbT['lop'], df_lop_dtbT['dtbToan'], marker='o', linestyle='dashed', color='green')
plt.xlabel('Lớp', size=15)
plt.ylabel('Điểm trung bình môn Toán', size=15)
plt.ylim(0,10)
#Hoặc plt.yticks(np.arrange(11))
plt.title('Biểu đồ điểm trung bình môn Toán của khối lớp 10', color='r')

plt.fill_between(df_lop_dtbT['lop'],df_lop_dtbT['dtbToan'], color = 'Slateblue', alpha=0.5) #Tô màu slateblue miền bên dưới biểu đồ đường

plt.show
```
